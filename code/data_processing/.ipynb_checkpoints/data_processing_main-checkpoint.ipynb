Crash Dataset

```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_columns = None
```

```python
crash_data_path = "/home/ruman/ruman_v/nexus/wright/intro_to_knowledge_engineering/Project/datasets/FARS/FARS2022NationalCSV/accident.csv"
print(crash_data_path)
```

```python
crash_data = pd.read_csv(crash_data_path, encoding='latin1')
```

```python
crash_data.columns
```

```python
crash_data['STATE'].unique()
```

```python
crash_data['FATALS'].unique()
```

```python
refractor_cols = ['ST_CASE','VE_TOTAL','PERSONS','WEATHERNAME','LGT_CONDNAME','WRK_ZONENAME', 'TYP_INTNAME','FATALS',]
refractor_crash_data = crash_data[refractor_cols]
refractor_crash_data.head()
```

```python
cleaned_data = refractor_crash_data.copy()
categorical_columns = ['WEATHERNAME', 'LGT_CONDNAME', 'WRK_ZONENAME', 'TYP_INTNAME']

for col in categorical_columns:
    cleaned_data[col] = cleaned_data[col].fillna("Unknown").str.replace(" ", "_")
```

```python
cleaned_data.info()
```

```python
cleaned_data.to_csv('crash_data.csv', index=False)
```
